# Simples tests on Iris datasets
This is a small notebook presenting basic machine learning algorithms applied on the Iris dataset.

```python
"""
This function takes a model as first argument, train it on iris dataset and displays its result.
"""
def show(clf):

    # Imports
    from sklearn import datasets
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import accuracy_score
    from mlxtend.plotting import plot_decision_regions
    import matplotlib.pyplot as plt

    # Make datasets
    iris = datasets.load_iris()
    X = iris.data[:, 0:2]
    y = iris.target
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3)

    # Train classifier
    clf.fit(Xtr, ytr)
    print('Accuracy: %.2f' % accuracy_score(yte, clf.predict(Xte)))

    # Plot decision regions
    plot_decision_regions(X, y, clf=clf, legend=2)
    plt.xlabel('sepal length [cm]')
    plt.ylabel('petal length [cm]')
    plt.show()
```

## Perceptron

```python
from sklearn.linear_model import Perceptron
show(Perceptron(tol=1e-3, random_state=0))
```

## SGD Classifier

```python
from sklearn.linear_model import SGDClassifier
show(SGDClassifier(max_iter=1000, tol=1e-3))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
show(LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial'))
```

## SVC

```python
from sklearn.svm import SVC
show(SVC(C=0.5, kernel='linear'))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
show(DecisionTreeClassifier(criterion="gini", max_depth=None, random_state=1))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
show(RandomForestClassifier(criterion="gini", n_estimators=10, random_state=1))
```

## K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
show(KNeighborsClassifier(n_neighbors=6, p=2, metric="minkowski"))
```
